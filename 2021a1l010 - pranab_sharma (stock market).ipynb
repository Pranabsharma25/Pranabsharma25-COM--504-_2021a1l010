```python
#store the data

df = pd.read_csv("KO.csv")

#Show the data

df 
```

```python
#Show and store the last row of data

actual_price=df.tail(1)

#show the last row

actual_price
```

```python
#Get all of the data expect the last row

df = df.head(len(df)-1)

#show the data set
df
```

```python
#create empty list

days = list()
abj_close_prices= list()
```

```python
#Get only the dates and the abjusted close prices

df_days= df.loc[:, 'Date']
df_abj_close= df.loc[:, 'Adj Close']
```

```python
#Create the indipendent data set (dates)

for day in df_days:
  days.append([int(day.split('-')[2])])

#Create the dependent dataset (Abj Close Prices)


for abj_close_price in df_abj_close:
  abj_close_prices.append(float(abj_close_price))




```

```python
#Show the days
days
```

```python
#show the abj close price

abj_close_prices

```

Support-Vector-Machine.png

 **SVR**
Supervised Machine Learning Models with associated learning algorithms that analyze data for classification and regression analysis are known as Support Vector Regression

images.png

```python
#Create 3 modles
x_train, x_test, y_train, y_test = train_test_split(days, abj_close_prices, random_state = 0, test_size=0.25)


lin_svr = SVR(kernel='linear',C=1)
lin_svr.fit(x_train, y_train)


poly_svr = SVR(kernel='poly', C= 1000.0, degree=2)
poly_svr.fit(x_train, y_train)


rbf_svr = SVR(kernel='rbf', C= 1000.0, gamma=0.85)
rbf_svr.fit(x_train, y_train)
```

```python
#plot the models

plt.figure(figsize=(16,8))
plt.scatter(days, abj_close_prices, color= 'black', label = 'Data')
plt.plot(days, rbf_svr.predict(days), color= 'green', label = 'RBF Model')
plt.plot(days, poly_svr.predict(days), color= 'orange', label = 'Polynomial Model')
plt.plot(days, lin_svr.predict(days), color= 'blue', label = 'Linear Model')
plt.xlabel('Days')
plt.ylabel('Abj Close Price ($)')
plt.legend()
plt.show()
```

```python
#Show the predicted price for the given day
dayinput=input("Enter the day you want to predict: ")
#day1 = [[30]]
day1 = [[dayinput]]
print('The RBF SVR predcted price:', rbf_svr.predict(day1))
print('The Linear SVR predcted price:', lin_svr.predict(day1))
print('The Polynomial SVR predcted price:', poly_svr.predict(day1))
```

```python
#Show the actual price

print('The actual price of the stock on day',dayinput,"is : 62.150002" )
```

**Accuracy of 3 kernals of SVR model**

```python
#Accuracy of linear model



score3 = lin_svr.score(x_test, y_test)
print(score3*(-1))
```

```python
#Accuracy of Polynomial model



score1 = poly_svr.score(x_test, y_test)
print(score1)
```

```python
#Accuracy of rbf model


score2 = rbf_svr.score(x_test, y_test)
print(score2)
```
